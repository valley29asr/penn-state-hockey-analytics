```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/penn_state_processed_stats_2024_2025.csv")

df.head()
df.info()
df.describe()

print("Total games:", len(df))
df["Win"].value_counts()
df["Win"].value_counts(normalize=True)*100  #win percentage

win_loss = df.groupby("Win").mean(numeric_only=True)
win_loss

comparison = df.groupby("Win").mean(numeric_only=True).T
comparison["Difference"] = comparison[1] - comparison[0]

comparison = comparison.sort_values(by = "Difference", ascending = False)

comparison

comparison.to_csv("../reports/win_loss_comparison.csv")
```
