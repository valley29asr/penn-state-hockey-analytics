# Exploratory Data Analysis

This notebook explores Penn State Men's Hockey statistics from the
2024–2025 season to identify differences between wins and losses.

```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv("../data/processed/penn_state_processed_stats_2024_2025.csv")
```

```python

df.head()
df.info()
df.describe()
```

```python


print("Total games:", len(df))
df["Win"].value_counts()
df["Win"].value_counts(normalize=True)*100  #win percentage
```

## Comparing Wins vs. Losses

```python

win_loss = df.groupby("Win").mean(numeric_only=True)
win_loss

comparison = df.groupby("Win").mean(numeric_only=True).T

comparison = comparison.drop(
    index=["WMT Game ID"],
    errors="ignore"
)

comparison["Difference"] = comparison[1] - comparison[0]

comparison = comparison.sort_values(
    by="Difference",
    ascending=False
)

comparison
```

## Export Results

```python

comparison.to_csv("../reports/win_loss_comparison.csv")
```
